```python
import pandas as pd
import numpy as np
```

```python
# Using glob module to list all files
import glob
```

```python
glob.glob("../data/sales-*-2014.xlsx")
```

```python
# Using glob to import each file and append to combine
all_data = pd.DataFrame()
for f in glob.glob("../data/sales-*-2014.xlsx"):
    df = pd.read_excel(f)
    all_data = all_data.append(df,ignore_index=True)
```

```python
all_data.describe()
```

```python
all_data.head()
```

```python
# Convert date column to date time object
all_data['date'] = pd.to_datetime(all_data['date'])
```

```python
# Combine more data. Read another file containing customer status by account
status = pd.read_excel("../data/customer-status.xlsx")
status
```

```python
# Doing left join to concatenate related data, similar to vlookup in excel
all_data_st = pd.merge(all_data, status, how='left')
all_data_st.head()
```

```python
# For account 737550
all_data_st[all_data_st["account number"]==737550].head()
```

```python
# Label all missing accounts as bronze
all_data_st['status'].fillna('bronze',inplace=True)
all_data_st.head()
```

```python
all_data_st[all_data_st["account number"]==737550].head()
```

```python
pd.__version__
```

```python
# Type cast status as category
all_data_st["status"] = all_data_st["status"].astype("category")
```

```python
all_data_st.head()
```

```python
all_data_st.dtypes
```

```python
#Current category ordering is alphabetical
all_data_st.sort_values(by=["status"]).head()
```

```python
 # setting ordering as olympic medal ordering
all_data_st["status"].cat.set_categories([ "gold","silver","bronze"],inplace=True)
```

```python
# Checking category ordering again to confirm
all_data_st.sort_values(by=["status"]).head()
```

```python
# Describing only status column
all_data_st["status"].describe()
```

```python
# Using groupby to calculate average values to see how top tier customers are performing compared to bottom
all_data_st.groupby(["status"])["quantity","unit price","ext price"].mean()
```

```python
all_data_st.groupby(["status"])["quantity","unit price","ext price"].agg([np.sum,np.mean, np.std])
```

```python
# Trying to understand why the volumes are skewed towards bronze customers
all_data_st.drop_duplicates(subset=["account number","name"]).iloc[:,[0,1,7]].groupby(["status"])["name"].count()
```
